## Image IO and editing with Pillow

### a) Implement the Image Loading and Saving Function

In this task, you will work with the Python Imaging Library (PIL), which is part of the Pillow package, to load and save images. The task will involve loading an image from the disk, performing a basic manipulation, and then saving the modified image back to the disk. You will also demonstrate how to convert images between different formats.

**Objective:**
- Learn how to load images using PIL.
- Save the manipulated images in different file formats.

**Requirements:**
1. **Load an Image**: Start by loading an image from your local file system. The image path should be passed as a parameter to your function.
2. **Save the Image**: Save the rotated image back to the disk in two formats:
   - As a PNG file with the postfix `_copy.png`.
   - As a JPEG file with the postfix `_copy.jpg`.

**Expected Output:**
- The script should demonstrate loading an image and saving it in two different formats. The paths to the saved images should be printed as output.

**Output from Example:**
```
Image saved as: path/to/your/image_copy.png
Image saved as: path/to/your/image_copy.jpg
```

````python
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
import numpy as np

# A random picture is downloaded and the path is defined to a variable image_path
# For all the tasks, this image path is called and same image is used
image_path = "/Users/aruneshsubramanians/Downloads/photo-1507608616759-54f48f0af0ee.webp"

# Function to visualize the images sequentially along with their respective labels
def display_image(images, titles):
    num_images = len(images)
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))
    for i in range(num_images):
        axes[i].imshow(images[i], cmap='gray')
        axes[i].set_title(titles[i])
        axes[i].axis('off')
    plt.tight_layout()
    plt.show()
````

````python
def image_loading_saving(image_path):
    # Open the image
    original_image = Image.open(image_path)
    
    # Rotate image by 90 degrees
    rotated_image = original_image.rotate(90)
    
    # Save the manipulated image as PNG
    png_output_path = image_path.replace('.png', '_copy.png')
    rotated_image.save(png_output_path)
    print("Image saved as:", png_output_path)
    
    # Save the manipulated image as JPEG
    jpeg_output_path = image_path.replace('.png', '_copy.jpg')
    rotated_image.save(jpeg_output_path)
    print("Image saved as:", jpeg_output_path)

image_loading_saving(image_path)
````

### b) Task Description: Image Manipulation with PIL


In this task, you will use the Pillow library to perform a series of image manipulation operations. The script will transform an image through a predefined sequence of manipulations, including rotation, color conversion, resizing, and brightness and contrast adjustments. This exercise provides a practical understanding of handling and modifying image data programmatically.

**Objective:**
- To apply multiple transformations to an image and understand the impact of each operation on the image data.

**Requirements:**
1. **Load an Image**: Import an image from your local file system. The path to the image should be parameterized.
2. **Rotate the Image**: Rotate the image 90 degrees clockwise.
3. **Convert to Grayscale**: Convert the color image to grayscale.
4. **Resize the Image**: Change the dimensions of the image to 128x128 pixels.
5. **Adjust Brightness**: Increase the brightness by 10%, and by a decrease of 10%.
6. **Adjust Contrast**: Increase the contrast by 10%, and by a decrease of 10%.

**Expected Output:**
- The script should sequentially perform the mentioned operations and show the manipulated image at the end of each task.

````python
def image_manipulation(image_path):
    # Load the image
    original_image = Image.open(image_path) 

    # Rotate the image 90 degrees clockwise
    rotated_image = original_image.rotate(-90)

    # Convert to grayscale
    grayscale_image = rotated_image.convert("L")

    # Resize to 128x128
    resized_image = grayscale_image.resize((128, 128))

    # Increase brightness by 10%
    brightness_enhancer = ImageEnhance.Brightness(resized_image)
    bright_image = brightness_enhancer.enhance(1.1)

    # Decrease brightness by 10%
    dark_image = brightness_enhancer.enhance(0.9)

    # Increase contrast by 10%
    contrast_enhancer = ImageEnhance.Contrast(dark_image)
    high_contrast_image = contrast_enhancer.enhance(1.1)

    # Decrease contrast by 10%
    low_contrast_image = contrast_enhancer.enhance(0.9)

    # Display images
    images = [original_image, rotated_image, grayscale_image, resized_image,
              bright_image, dark_image, high_contrast_image, low_contrast_image]
    titles = ["Original", "Rotated", "Grayscale", "Resized",
              "Bright", "Dark", "High Contrast", "Low Contrast"]
    display_image(images, titles)

image_manipulation(image_path)
````

## 2. Image Tensor Manipulation with numpy 

In this task, you will work with image data represented as NumPy arrays. You will perform basic image manipulation operations using NumPy, such as flipping, rotating, and resizing images. This exercise will help you understand how to manipulate image data using NumPy arrays and visualize the results.

**Objective:**
- To perform basic image manipulation operations using NumPy arrays.

**Requirements:**

1. **Load an Image**: Load an image from your local file system using Pillow and convert it to a NumPy array.
2. **Rotate the Image**: Rotate the image 90 degrees clockwise
3. **Convert to Grayscale**: Convert the color image to grayscale.
4. **Increase Image Brightness**: Increase the brightness of the image by 10%.
5. **Decrease Image Brightness**: Decrease the brightness of the image by 10%.
6. **Increase Image Contrast**: Increase the contrast of the image by 10%.
7. **Decrease Image Contrast**: Decrease the contrast of the image by 10%.

**Expected Output:**
- The script should sequentially perform the mentioned operations and display the manipulated image at the end of each task.


````python
def image_manipulation_numpy(image_path):
    # Open an image file and convert to numpy array
    pil_image = Image.open(image_path)
    image_np = np.array(pil_image)

    # Rotate the image 90 degrees clockwise
    rotated_image = np.rot90(image_np, k=-1)

    # Convert to grayscale
    grayscale_image = np.dot(rotated_image[..., :3], [0.2989, 0.5870, 0.1140])
                         
    # Increase the brightness by 10%
    bright_image = np.clip(grayscale_image * 1.1, 0, 255).astype(np.uint8)

    # Decrease the brightness by 10%
    dark_image = np.clip(bright_image * 0.9, 0, 255).astype(np.uint8)

    # Increase the contrast by 10%
    mean_value = np.mean(dark_image)
    high_contrast_image = np.clip((dark_image - mean_value) * 1.1 + mean_value, 0, 255).astype(np.uint8)

    # Decrease the contrast by 10%
    low_contrast_image = np.clip((high_contrast_image - mean_value) * 0.9 + mean_value, 0, 255).astype(np.uint8)

    # Display images
    images = [image_np, rotated_image, grayscale_image, 
              bright_image, dark_image, high_contrast_image, low_contrast_image]
    titles = ["Original", "Rotated", "Grayscale", 
              "Bright", "Dark", "High Contrast", "Low Contrast"]   
    display_image(images, titles)

image_manipulation_numpy(image_path)
````
